```python
# Optimization technique
# (1). Gradient Descent ---- > this is the technique to reduce loss function(Distance from
# y_test to y_pred).

# We know that Equation of Simple Line :::
# y = mx+b . . . .. (1)

# Loss = (y_act - y_pred)
# So , we will calculate Loss w.r.t. Mean Squared Error
# L = 1/n * (y_act - y_pred) ** 2           ..... (2)

# where n = n(features) = 1
# L = (y_act - y_pred) ** 2         .... (3)
# L = (y-(mx+b) ) ** 2              
# L = (y -mx -b) ** 2           .... (4)


# Now, we will calculate differentiation of Loss w.r.t. Intercept(b)
# dL/db = 2(y-mx-b).[dy/db - dmx/db - db/db]
# dL/db = 2(y-mx-b).[0-0-1]
# dL/db = -2(y-mx-b)            ...... (5)

# Now, we will calculate differentiation of Loss w.r.t. Weights(m)
# dL/dm = 2(y-mx-b). [dy/dm - dmx/dm - db/dm]
# dL/dm = 2(y-mx-b).[0-x-0]
# dL/dm = -2x(y-mx-b)       ..... (6)



### learning rate = constant = 0.001(In general)


# b_new = b_old - learning_rate * dL/db

# m_new = m_old - learning_rate * dL/dm

# if learning_rate >>>>> 0  (like LR = 1,8,25,etc)    ---- > Gradient Exploading
# if learning_rate ~ 0 (LR tending to 0 like LR = 0.0000005) ---- > Vanishing Gradient

# So , the final Equation of New Line will be ::
# y = m_new * x + b_new

### learning_rate = constant = 0.001(In general)





```
