# svm is a supervised ml model mainly used for classificatrion but can also be used for regression  
# # the main objective is to find the best boundries that separates different classes
# this boundry is called hyper plain
# SVM tries to maximise the margin between classes 

# WOrking :- 
    # take traning data then find possible separating boundries 
    #  then calculate margine for each boundry 
    #  THEN select the boundry with max margin 
    #  then it use that boundry for pridiction on new data

#  imp parameter :- 
    # c parameter :- it controlls the misclassification . 
            # small c is larger margin but has more errors
            # large c means small margin but fewer errors
    
    #  gamma parameter
        # it is mainly used in RBF karnel 
        # smaller gamma means smoother boundry and large gamma tends to have complex boundries
    
    

SVM vs other models :- 
    --> it is better than logistic regression when data is complex 
    --> it is better than DT when margin baised separation is usefull 
    --> it is sometimes slower than RF on huge dataset
    --> and it is often used when accuracy is more important than interpritablity 
    

Types of SVM :- 
    --> linear SVM :- it is used when data is linearly separable
    --> non linear SVM :- it is used when data is not separable linearly and it uses karnel trick 
    --> it uses karnel like linear karnel , polynomial karnel , RVF KARNEL AND sigmoid karnel 

Important karnel selection rules :- 
     Linear karnel --> simple and fast
     RVF karnel --> powerful and flexible
     Polinomial karnel --> useful for curved boundries
     sigmoid karnel --> it is not very commenly used 

     if c is very high then model tries to classify all training points correctly and risk of overfitting increases 
     if c is very low then model allows some errors and  better generalisation may happen 
     
     if gamma is very high then the model creates very complex decision boundry and it may results in overfitting 
     if gamma is very low then the model creates smooth decision boundy and it may result in underfitting 

     SVM does not handle missing values directly 

Support vectors are the only points that influence the hyper plane 
other distant point do not affect the decision boundry so much 
SVM focus on maximising margin while algorythems like logistic regression focus on probablity estimation 



```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\codings\datasets\Social_Network_Ads - Social_Network_Ads.csv")
```

```python
df.head(5)
```

```python
df = df.dropna()
```

```python
df_cat = df.select_dtypes(include='object').columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
for x in df_cat:
    lb = LabelEncoder()
    df[x] = lb.fit_transform(df[x])
```

```python
x = df.drop(columns=['Purchased'])
y = df['Purchased']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear', C=1.0, gamma='scale')
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)

grid.fit(x_train, y_train)
```

```python


from sklearn.metrics import classification_report, accuracy_score

grid_predictions = grid.predict(x_test)

print(classification_report(y_test, grid_predictions))
```
