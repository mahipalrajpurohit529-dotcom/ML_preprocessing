```python
import pandas as pd
```

```python
# insurance data all regression models
```

```python
df = pd.read_csv(r"C:\codings\datasets\insurance - insurance.csv")
```

```python
df
```

```python
df = df.dropna()
```

```python
df_cat = df.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
for x in df_cat:
    df[x] = lb.fit_transform(df[x])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['charges'])
y = df['charges']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import r2_score
```

```python
# linear regression 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lgr = LinearRegression()
```

```python
lgr.fit(x_train,y_train)
```

```python
y_pred = lgr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
# decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train )
```

```python
y_pred = dtr.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

```python
# random forest 
```

```python

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train , y_train)
```

```python
y_pred = rfr.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

```python
# SVM
from sklearn.svm import SVR
sv = SVR(kernel='linear', C=1.0, gamma='scale')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
sv.fit(x_train_sc,y_train)
```

```python
y_pred=sv.predict(x_test_sc)
```

```python
r2_score(y_test,y_pred)
```

```python
# KNN
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
