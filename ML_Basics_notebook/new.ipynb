```python
# pipeline and column transformer for encoding and standerization :-
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.info()
```

```python
transformer = ColumnTransformer(transformers=[
('tnf',OneHotEncoder(sparse_output=False, drop='first' ), ['sex' , 'smoker' , 'day' , 'time'])
],remainder='passthrough')
```

```python
categorical_transformer = Pipeline(steps=[
('ct', transformer),
('stnd', StandardScaler())
])
```
