```python
df_cat = df.select_dtypes(include= "category").columns.to_list()
```

```python
### to convert all categorial data into numerical by the help of for loop
for i in df_cat:
  uni = df[i].unique().tolist()
  num = 0
  empty = {}
  for j in uni :
    empty[j] = num
    num +=1
  df[i]=df[i].map(empty)
```

TRAIN TEST SPLIT :-

```python
x = df.drop(columns='time')
```

```python
y = df['time']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,
    test_size=0.3,
    random_state=4
)
```
