HERE I TRY TO NORMALISE AND STANDERIZE THE GIVE DATA WHILE ALSO USING THE LABLE ENCODER

```python
# STEPS TO DO :- 
    # fill null values 
    # remove the catagorical columns having more than 2,3 unique values 
    # convert the remaining catagorical columns into numerical columns (use lable encoder)
    # split the data using train test split
    # normalise
    # standerize
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\asus\Downloads\Attrition - Attrition.csv")
```

```python
df.isna().sum().sum()       # null values check
```

```python
df_Cat = df.select_dtypes(include='object')
df_Cat.nunique()
```

```python
df = df.drop(columns=['EducationField','JobRole','MaritalStatus'])          # droping unwanted columns
df_Cat = df_Cat.drop(columns=['EducationField','JobRole','MaritalStatus'])
```

```python
from sklearn.preprocessing import LabelEncoder          # importing lable encoder class
```

```python
lb = LabelEncoder()             # creating objecct of lable encoder
```

```python
df_Cat_list = df.select_dtypes(include='object').columns.tolist()       # creating list of column names
```

```python
for x in df_Cat_list:
    df[x]= lb.fit_transform(df[x])              # converting all catagorical columns to numerical using the lable encoder 
```

```python
# df.info()
```

```python
# NOW TRAIN TEST SPLIT :-
```

```python
x = df.drop(columns=['Attrition'])
```

```python
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,
    test_size=0.2,
    random_state=40
)
```

```python
# NOW STANDAZATION :-
```

```python
from sklearn.preprocessing import StandardScaler            # importing standered scaler class
```

```python
sc = StandardScaler()           # creaating object
```

```python
x_train_sc = sc.fit_transform(x_train)          # standerizing the x_train data
```

```python
x_train_new = pd.DataFrame(x_train_sc, columns=x_train.columns)         # coverting the standarized data into a DF
```

```python
np.round(x_train_new.describe(),2)      # rounding for visualisation because there are a lot of digits after the decimals
```

```python
# NORMALISATION(minmaxscaling) :- 
```

```python
from sklearn.preprocessing import MinMaxScaler        # imorting the minmaxscaler class
```

```python
mn = MinMaxScaler()         # creating object of the class
```

```python
x_train_new = mn.fit_transform(x_train_new)
```

```python
x_train_new = pd.DataFrame(x_train_new,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
