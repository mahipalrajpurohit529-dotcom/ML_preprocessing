```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\codings\datasets\titanic - titanic.csv")
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df.isna().mean()*100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.21 and df[var].isnull().mean() > 0]
cols
```

```python
df['Survived'].value_counts()
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# original data
df['Age'].hist(bins=50, ax=ax, density=True, color='red' )

# data after cca, the argument alpha makes the color transparent, so we can
# see the overlay of the 2 distributions
new_df['Age'].hist(bins=50, ax=ax, color='green', density=True, alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# original data
df['Fare'].hist(bins=50, ax=ax, density=True, color='red' )

# data after cca, the argument alpha makes the color transparent, so we can
# see the overlay of the 2 distributions
new_df['Fare'].hist(bins=50, ax=ax, color='green', density=True, alpha=0.8)
```
