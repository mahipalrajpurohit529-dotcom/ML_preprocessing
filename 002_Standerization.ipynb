```python
# NORMAL DISTRIBUTION :- 
    # it is a way to convert our data into smaller size without changing its central tendencies
    # we make sure half the data is on left side and 
    # the other half is on the right side of the central values
```

```python
# Example for understanding:-

# Age = 1,2,3,4,5
# mean= (1+2+3+4+5)/5 = 15/5 = 3
# std = sqrt(sum(xi-mean)**2/N)
# sqrt((1-3)**2 + (2-3)**2 + (3-3)**2 + (4-3)**2 + (5-3)**2)/5
# sqrt(4+1+0+1+4)/5 = sqrt(2) = 1.2  = 1

# standrd deviation  = (x_current - mean)/standard_deviation
# Age' = (1-3)/1 = -2  , (2-3)/1 = -1 , (3-3)/1 = 0 , (4-3)/1 = 1 , (5-3)/1 = 2
# Age' = -2,-1,0,1,2
# mean = (-2-1+0+1+2)/5 = 0
# std = 1
```

```python
# for converting our data into Normal Distribution we will use StandardScaler
# this process is called as standarization
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()       # creating object of the class
```

```python
x_train_sc = sc.fit_transform(x_train)      # performing standarization with a class called fit tranform
    # but this results an array and we need a DATA FRAME
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x_train.columns)  # converting it into df 
```
