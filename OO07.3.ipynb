```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['charges']),df['charges'],test_size=0.2,random_state=34)
```

```python
processed = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(drop='first',sparse_output=False),['sex','smoker','region'])
],remainder='passthrough')
```

```python
z=processed.fit_transform(x_train)
```

```python
y = processed.get_feature_names_out()
```

```python
newDF = pd.DataFrame(z,columns=y)
```
