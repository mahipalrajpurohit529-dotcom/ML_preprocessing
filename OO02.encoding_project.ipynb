```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
# 1st :- lable encoder
```

```python
df = sns.load_dataset('Tips')
df = df.dropna()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['sex'] = lb.fit_transform(df['sex'])
df['smoker'] = lb.fit_transform(df['smoker'])
df['day'] = lb.fit_transform(df['day'])
df['time'] = lb.fit_transform(df['time'])
```

```python
df
```

```python
# 2nd :- OrdinalEncoderf
```

```python
df = sns.load_dataset('Tips')
df = df.dropna()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = df.drop(columns=['total_bill','tip','size'])
```

```python
oe = OrdinalEncoder( categories=[
    ['Female','Male'],
    ['No','Yes'],
    ['Sun', 'Sat', 'Thur', 'Fri'],
    ['Dinner', 'Lunch']
])
```

```python
cpoe = df
```

```python
new_oe = oe.fit_transform(df)       
# this works but the result is an array and we need df
```

```python
new_oe = pd.DataFrame(new_oe,columns=cpoe.columns)
```

```python
new_oe.sample(3)
```
